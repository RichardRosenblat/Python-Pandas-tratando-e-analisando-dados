```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('../data/aluguel.csv', sep=';')
```

```python
tipos = pd.DataFrame(dataset['Tipo'].drop_duplicates())
tipos.index = range(tipos.index.shape[0])
tipos.head()
```

```python
value_by_type_series = dataset.groupby('Tipo').mean()['Valor']
value_by_type = pd.DataFrame(value_by_type_series)
value_by_type.rename(columns = {"Valor": "Valor Avg"}, inplace = True)
value_by_type
```

```python
value_by_type.plot.bar(y="Valor Avg")
```

```python
value_by_type.drop(index=['Prédio Inteiro'], inplace=False).plot.bar(y="Valor Avg")
```

```python
value_by_type.drop(index=['Prédio Inteiro','Indústria'], inplace=False).plot.bar(y="Valor Avg")
```

```python
value_by_type.drop(index=['Prédio Inteiro','Indústria','Galpão/Depósito/Armazém'], inplace=False).plot.bar(y="Valor Avg")
```
