```python
import matplotlib.pyplot as plt
```

```python
meses = ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
valores = [105235, 107697, 110256, 109236, 108859, 109986, 110256, 110487, 110239, 110339, 109994, 109974]

# valores[:6]
# meses[:6]
```

```python
plt.figure(figsize=(12,5))
plt.plot(meses,valores)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(meses,valores)
plt.ylim(104000, 114000)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.title('Faturamento no primeiro semestre de 2017')
plt.ylim(104000, 114000)
plt.plot(meses,valores)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.title('Faturamento no primeiro semestre de 2017')
plt.ylim(104000, 114000)
plt.xlabel('Meses')
plt.ylabel('Faturamento em R$')
plt.plot(meses, valores, label='2021', linestyle="-.")
valores_copy = valores.copy()
valores_copy.reverse()
plt.plot(meses, valores_copy, label='2022', linestyle="-")
plt.legend()
plt.show()
```
