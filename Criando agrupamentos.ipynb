## Criando agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('./data/aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

https://pandas.pydata.org/docs/reference/frame.html#computations-descriptive-stats

```python
dados['Valor'].mean()
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
# grupo_bairro.groups
```

```python
for bairro, imoveis in grupo_bairro:
    print(f'{bairro} -> {round(imoveis["Valor"].mean(), 2)}')
```

```python
dados.groupby('Bairro').mean().round(2)
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatisticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

## Visualizando dados

```python
grupo_bairro['Valor'].std().plot.bar(color='green')
```

```python
import matplotlib.pyplot as plt
# Para alterar o tamanho de todos os graficos é utilizado:
# plt.rc('figure', figsize = (40,20))
```

```python
dados_completos = pd.read_csv('./data/aluguel_residencial.csv', sep=';')
grupo_bairro_completos = dados_completos.groupby('Bairro')
```

```python
fig = grupo_bairro_completos['Valor'].mean().plot.bar(color='green')
fig.set_ylabel('Valor do aluguel')
fig.figure.set_figwidth(40)
fig.figure.set_figheight(10)
fig.set_title('Valor Médio do aluguel por Bairro (Todos)', {'fontsize': 44})
plt.show()
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color='green')
fig.set_ylabel('Valor do aluguel')
fig.set_title('Valor Médio do aluguel por Bairros (Seleção)')
plt.show()
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color='green')
fig.set_ylabel('Valor do aluguel')
fig.set_title('Valor Máximo do aluguel por Bairro (Seleção)')
plt.show()
```
