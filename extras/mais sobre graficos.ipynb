```python
# %matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
# Seleciona uma amostra aleatoria e redefine os indexes para ficarem em ordem
amostra = dados['Valor'].sample(100)
amostra.index = range(amostra.shape[0])

# Agrupa valor por tipo de imóvel
grupo = dados.groupby('Tipo')['Valor']

# Separa indices e valores médios dos grupos
indices_grupo = grupo.mean().index
valores_grupo = grupo.mean().values
```

```python
# Cria figura
area = plt.figure(figsize=(20, 13))
```

```python
# Cria graficos
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
# Configura grafico 1
g1.scatter(dados['Valor'],dados['Area'])
g1.set_title('Valor X Area')

# Configura grafico 2
g2.hist(dados['Valor'])
g2.set_title('Histograma Valor')

# Configura grafico 3
g3.plot(amostra)
g3.set_title('Amostra Aleatória de Valores')

# Configura grafico 4
g4.bar(indices_grupo, valores_grupo)
g4.set_title('Valor Médio por Tipo')
```
